```python

import numpy as np
import numpy.random as npr
import pandas as pd

def bootstrap(invec):
    idx = npr.randint(0, len(invec), len(invec))
    return [invec[i] for i in idx]

def estimatePi0(p, numBoot=100, numLambda=100, maxLambda=0.95):
    p.sort()
    n=len(p)
    lambdas=np.linspace(maxLambda/numLambda,maxLambda,numLambda)
    Wls=np.array([n-np.argmax(p>=l) for l in lambdas])
    pi0s=np.array([Wls[i] / (n * (1 - lambdas[i])) for i in range(numLambda)])
    minPi0=np.min(pi0s)
    mse = np.zeros(numLambda)
    for boot in range(numBoot):
        pBoot = bootstrap(p)
        pBoot.sort()
        WlsBoot =np.array([n-np.argmax(pBoot>=l) for l in lambdas])
        pi0sBoot =np.array([WlsBoot[i] / (n *(1 - lambdas[i])) for i in range(numLambda)])
        mse = mse + np.square(pi0sBoot-minPi0)
    minIx = np.argmin(mse)
    return pi0s[minIx]

# The input to this function is tupples of p-values and analyte names, e.g. (p,coord)
def qvalues(pvalues):
    m=len(pvalues)
    pvalues.sort()
    pi0 = estimatePi0([p for p,coord in pvalues])
    num_p, qs = 0.0, []
    for p,coord in pvalues:
        num_p += 1.0
        q = pi0*p*m/num_p
        qs.append((q,p,coord))
    qs.reverse()
    old_q=1.0
    for ix in range(len(qs)):
        q = min(old_q,qs[ix][0])
        old_q = q
        qs[ix] = (q,qs[ix][1],qs[ix][2])
    qs.reverse()
    return qs
```

```python
## Intializing the path and required packages for data analysis 

import os
import pandas as pd
import numpy as np

os.chdir("/Users/salendrapradh/Documents/prj_algo_bioinfo/data")

df = pd.read_excel("MSB-12-901-s009.xlsx",skiprows=list(range(6))+list(range(15,18)),parse_cols=[0]+list(range(3,318)),header=0,index_col=0, encoding='sys.getfilesystemencoding()')
weight = pd.read_excel ("MSB-12-901-s009.xlsx",skiprows=list(range(6))+list(range(8,755)),parse_cols=[0]+list(range(3,318)),header=0,index_col=0, encoding='sys.getfilesystemencoding()')

print(df.shape)
print(weight.shape)
```

```python
weight.loc["Subject_id",:] = [int(x.split("_")[1])for x in weight.columns.values]
weight.loc["Time",:] = [int(x.split("_")[2])for x in weight.columns.values]
from_w = ['_{0:02d}_{1:d}'.format(int(i),1) for i in weight.loc["Subject_id"]]
to_w = ['_{0:02d}_{1:d}'.format(int(i),int(j)) for i,j in zip(weight.loc["Subject_id"],weight.loc["Time"])]
from_w = [s.replace('_01', '_1') for s in from_w]
from_w = [s.replace('_44_1', '_44_2') for s in from_w]
to_w = [s.replace('_01', '_1') for s in to_w]

delta_w = [weight.loc["Weight",to] - weight.loc["Weight",fr] for to,fr in zip(to_w,from_w)]
weight.loc["dW",:] = delta_w
df.dropna(inplace=True, axis =0)
df_2 = pd.concat([weight,df])
df_2.shape
```

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols
from statsmodels.stats.anova import anova_lm
header_rows = ["Subject_id", "Time", "Weight"]
pv,sub_var = [],{}
for name,values in df_2.iterrows():
    if name in header_rows:
        continue
    subframe = df_2.loc[header_rows+[name],:].T.copy()
    renam = {name:'Expression'}
    subframe.rename(columns=renam,inplace=True)
    formula = 'Expression ~ C(Subject_id) + C(Time)'
    model = ols(formula,subframe).fit()
    sub_var[name] = model.params.filter(regex=("C.*Subject_id.*")).var()
    table = sm.stats.anova_lm(model, typ=2)
    pvalue = table.loc['C(Time)'][-1]
    pv.append((pvalue,name))
qv = qvalues(pv)
```

```python
import matplotlib.pyplot as plt
plt.plot([q[0] for q in qv],list(range(len(qv))),'b-')
plt.ylabel("Number accepted p-values")
plt.xlabel("q-value")
plt.xlim(0,0.1)
plt.ylim(0,200)
plt.show()
```

```python

plt.plot([sub_var[q[2]] for q in qv],[-np.log10(q[0]) for q in qv],'b.')
plt.ylabel("-log of q-values")
plt.xlabel("variance among subjects")
plt.xlim(.0,0.3)
plt.show()
```

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols
from statsmodels.stats.anova import anova_lm
header_rows = ["Subject_id", "Time", "Weight","dW"]
pv,sub_var = [],{}
for name,values in df_2.iterrows():
    if name in header_rows:
        continue
    subframe = df_2.loc[header_rows+[name],:].T.copy()
    renam = {name:'Expression'}
    subframe.rename(columns=renam,inplace=True)
    formula = 'Expression ~ C(Subject_id) + dW'
    model = ols(formula,subframe).fit()
    sub_var[name] = model.params.filter(regex=("C.*Subject_id.*")).var()
    table = sm.stats.anova_lm(model, typ=2)
    pvalue = table.loc['dW'][-1]
    pv.append((pvalue,name))
qv = qvalues(pv)
```

```python
plt.plot([q[0] for q in qv],range(1,len(qv)+1),'b-')
plt.ylabel("Number accepted p-values")
plt.xlabel("q-value")
plt.xlim(0,0.1)
plt.ylim(0,200)
plt.show()
```
